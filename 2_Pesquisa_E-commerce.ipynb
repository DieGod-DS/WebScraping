### Importando bibliotecas

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## 1. Utilizando produtos do Mercado Livre para estudo

```python
url_base = 'https://lista.mercadolivre.com.br/'

product_name = input('Produto desejado')

response = requests.get(url_base + product_name)

site = BeautifulSoup(response.text, 'html.parser')

products = site.findAll('div', attrs={'class': 'andes-card ui-search-result ui-search-result--core andes-card--flat andes-card--padding-16'})

for product in products: 
    # titulo do produto
    title = product.find('h2', attrs={'class':'ui-search-item__title'})

    # link do produto
    link = product.find('a', attrs={'class':'ui-search-item__group__element ui-search-link__title-card ui-search-link'})

    # valor do produto
    real = product.find('span', attrs={'class':'andes-money-amount__fraction'})
    cents = product.find('span', attrs={'class':'andes-money-amount__cents andes-money-amount__cents--superscript-24'})

    # print(product.prettify())
    print('Título do produto:', title.text)
    print('Link do produto:', link['href'])
    
    if (cents):
        print(f'Preço do produto R$', real.text + ',' + cents.text)
    else:
        print(f'Preço do produto R$', real.text)
    print('\n\n')
```
